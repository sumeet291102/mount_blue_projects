```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

asean_countries = ['Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam']
saarc_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']
un_data = pd.read_csv('./resources/population-estimates.csv')
```

```python
un_data.head()
```

## India population over years - Bar Plot

```python
india_population_over_years = {}

with open('./resources/population-estimates.csv') as file_obj:
    countries = csv.DictReader(file_obj)

    for country in countries:
        if country['Region'] != 'India':
            continue

        india_population_over_years[country['Year']] = india_population_over_years.get(country['Year'], 0) + float(country['Population'])

plt.barh(list(india_population_over_years.keys()), list(india_population_over_years.values()))
plt.show()
```

## For the year 2014. Bar Chart of population of ASEAN countries

```python
asean_countries_population_in_2014 = {}

with open('./resources/population-estimates.csv') as file_obj:
    countries = csv.DictReader(file_obj)

    for country in countries:
        if (country['Region'] not in asean_countries) | (country['Year'] != '2014'):
            continue

        asean_countries_population_in_2014[country['Region']] = asean_countries_population_in_2014.get(country['Region'], 0) + float(country['Population'])

plt.barh(list(asean_countries_population_in_2014.keys()), list(asean_countries_population_in_2014.values()))
plt.show()
```

## Over the years, TOTAL population of SAARC countries

```python
saarc_countries_total_population_over_the_years = {}

with open('./resources/population-estimates.csv') as file_obj:
    countries = csv.DictReader(file_obj)

    for country in countries:
        if country['Region'] not in asean_countries:
            continue

        saarc_countries_total_population_over_the_years[country['Year']] = saarc_countries_total_population_over_the_years.get(country['Year'], 0) + float(country['Population'])

plt.barh(list(saarc_countries_total_population_over_the_years.keys()), list(saarc_countries_total_population_over_the_years.values()))
plt.show()
```

## Grouped Bar Chart - ASEAN population vs. years

```python
asean_countrie_un_data=un_data[(un_data['Year']<=2014) & (un_data['Year']>=2004) & (un_data['Region'].isin(asean_countries))]
pivoted_data=asean_countrie_un_data.pivot_table(index='Year', columns='Region', values='Population').reset_index()
pivoted_data.plot(x='Year', 
        kind='bar', 
        stacked=False, 
        title='Grouped Bar Graph with dataframe') 
```
